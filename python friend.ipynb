```python
def read_csv(input):
    a = open(input).read()
    final_list = []
    string_list = a.split('\n')[1::]
    for i in string_list:
        int_fields = []
        string_fields = i.split(',')
        for n in string_fields:
            int_fields.append(int(n))
        final_list.append(int_fields)
    return final_list
```

```python
class Dataset:
    def __init__(self, data):
        self.header = data[0]
        self.data = data[1:]

    def column(self, label):
        if label not in self.header:
            return None
        index = 0
        for idx, value in enumerate(self.data):
            if value == label:
                index = idx
        data_col = []
        for i in self.data:
            data_col.append(i[index])
        return data_col
```

```python
class Dataset:
    def __init__(self, data):
        self.header = data[0]
        self.data = data[1:]
    
    def __str__(self):
        return str(self.data[:10])
    
    def column(self, label):
        if label not in self.header:
            return None
        
        index = 0
        for idx, element in enumerate(self.header):
            if label == element:
                index = idx
        
        column = []
        for row in self.data:
            column.append(row[index])
        return column
    
    def count_unique(self, label):
        unique_results = set(self.column(label))
        count = len(unique_results)
        return count
```
